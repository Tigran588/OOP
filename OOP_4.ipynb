```python
from abc import ABC, abstractmethod

# abstract classes for course assignments

class Assignments(ABC):
    pass

class BooksToRead(Assignments):
    def __init__(self) -> None:
        super().__init__()
        self._books_to_read = []

    def add_books(self, *books: str):
        for book in books:
            self._books_to_read.append(book)

class PracticTasks(Assignments):
    def __init__(self) -> None:
        super().__init__()
        self.__practic_tasks = []

    def add_books(self, *tasks: str):
        for task in tasks:
            self.__practic_tasks.append(task)

class SuplementaryAssignments(Assignments):
    def __init__(self) -> None:
        super().__init__()
        self._suplementary_assignmnets = []

    def add_books(self, *assignments: str):
        for i in assignments:
            self._suplementary_assignmnets.append(i)

class Student:
    pass

class Course(ABC):
    def __init__(self, name: str, instructor: str, content: str) -> None:
        super().__init__()
        self._name = name
        self._instructor = instructor
        self._content = content
        self._students = []
        self._assignments = []

    def add_student(self, student: Student):
        self._students.append(student)

    def add_assignment(self, assignment: Assignments):
        self._assignments.append(assignment)
        for st in self._students:
            st.notify()
            st.add_assignment(assignment)

    def get_info(self):
        print(f"Course name: {self._name}, instructor is {self._instructor}")


class GraduateCourse(Course):
    def get_info(self):
        print("Graduate Course")
        super().get_info()

class UndergraduateCourse(Course):
    def get_info(self):
        print("Undergraduate Course")
        super().get_info()

class Person(ABC):
    def __init__(self, name: str, contact_info: str) -> None:
        super().__init__()
        self._name = name
        self._contact_info = contact_info


class Student(Person):
    def __init__(self, name: str, contact_info: str) -> None:
        super().__init__(name, contact_info)
        self._cources = []
        self._assignments = []
        self._all_assignments = 0
        self._assignments_done = 0

    def enroll_course(self, course: Course):
        self._cources.append(course)
        course._students.append(self)
        print(f"{self._name} enrolled course: ", end="")
        course.get_info()

    def notify(self):
        print("There is new assignment")

    def add_assignment(self, assignment: Assignments):
        self._assignments.append(assignment)
        self._all_assignments += 1

    def complete_assignment(self, assignment: Assignments):
        if assignment in self._assignments:
            self._assignments_done += 1
            self._assignments.remove(assignment)
        else:
            "You have not such an assignment"

    def view_progress(self):
        print(f"You have done {self._assignments_done} assignment of {self._all_assignments}. There is {len(self._assignments)} assignmnets to do")

class Professor(Person):
    def create_graduate_course(self, name, instructor, content):
        return GraduateCourse(name, instructor, content)

    def create_undergraduate_course(self, name, instructor, content):
        return UndergraduateCourse(name, instructor, content)

if __name__ == "__main__":
    professor = Professor("Maria", "012 345 678")
    st1 = Student("St1", "smth@gmail.com")
    st2 = Student("St2", "098 765 432")

    course1 = professor.create_graduate_course("CS", professor._name, "Hard course")
    course2 = professor.create_graduate_course("Math", professor._name, "Interesting course")

    st1.enroll_course(course1)
    asign1 = BooksToRead().add_books("OOAD")
    course1.add_assignment(asign1)

    st1.view_progress()

    st1.complete_assignment(asign1)

    st1.view_progress()
```
