```python

##1##
from abc import ABC,abstractmethod

class Account(ABC):
  def __init__(self,account_type,balance,interest_rate):
    self.account_type = account_type
    self.balance = balance
    self.interest_rate = interest_rate

  @abstractmethod
  def withdraw(self,ammount):
    pass

  def deposit(self,ammount):
    pass

  def calc_interest(self):
    return self.balance * self.interest_rate

class Savings(Account):
  def __init__(self,balance):
    super().__init__('saving',balance,0.02)

  def withdraw(self,ammount):

    if self.balance >=  ammount:
      self.balance -=  ammount
    else:
      print('ammount must be lower then balance ')

  def deposit(self,ammount):
    self.balance += self.ammount

class Checking(Account):
  def __init__(self,balance):
    super().__init__('checking',balance, 0)

  def withdraw(self,ammount):
    if self.balance >= ammount:
      self.balance -= ammount
    else:
        print('ammount must be lower then balance ')

  def deposit(self,ammount):
    self.balance += ammount


class Customer:
  def __init__(self,name,address,contact):
    self.name = name
    self.address = address
    self.contact = contact


class Transaction:
  def __init__(self,account_inv,transaction_type,ammount):
    self.account_inv = account_inv
    self.transaction_type = transaction_type
    self.ammount = ammount


  def execute(self):
    if self.transaction_type == 'deposit':
      self.account_inv.deposit(self.ammount)
    if self.transaction_type == 'withdraw' :
      self.account_inv.withdraw(self.ammount)

c1 = Customer("Tigran",'Yerevn','Tarikon@mail.ru')
c2 = Customer('Garo','Artashat','mar@rambler.ru')

check = Checking(5000)
save = Savings(3000)

T1 = Transaction(check,'deposit',1000)
T1.execute()

T2 = Transaction(save,'withdraw',3001)
T2.execute()
save.balance
```
